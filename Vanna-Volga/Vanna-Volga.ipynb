# 1. Basic Concepts

## 1.1 Construct Riskless Portfolio

Assume $S_t$ follows Geometric Brownian Motion with stochastic but strike-independent implied volatility:
$$dS_t = (r_d - r_f)S_tdt + \sigma_t S_t dZt \tag{1.1}$$

In FX market, when we say BS, we mean：
$$O = \phi e^{-r_f\tau} S_t N(d_1) - \phi e^{-r_d\tau} K N(d_2) \tag{1.2}$$

where $\phi$ is just a indication of call and put (it doesn't matter actually), $\phi = \pm 1$

Similar to equity options, we construct riskless portfolio. 

But besides hedging delta, we hedge vanna and volga as well:
$$ d\pi = dO - \Delta dS - \sum_{i = 1}^3 x_idC_i \tag{1.3}$$

where $C_i$ is 75∆ call (or 25∆ put), ATM call (or ATM put), 25∆ call (or 75∆ put).

Their corresponding strikes are further denoted by $K_i, i = 1, 2, 3$, $K_1 < K_2 < K_3$ (which is equivalent to writing $K1 = K_{25P} , K2 = K_{ATM}, K3 = K_{25C}$ and the market-implied volatilities associated with $K_i$ are denoted by $σ_i$.

Exactly the same as deriving BS solutions, we use Ito Lemma to expand our portfolio:

$$d\pi = \frac{\partial O}{\partial S}dS + \frac{\partial O}{\partial t}dt + \frac{\partial O}{\partial \sigma}d\sigma + \\ \frac{1}{2}\{\frac{\partial^2 O}{\partial S^2}(dS)^2 + \frac{\partial^2 O}{\partial t^2}(dt)^2 + \frac{\partial^2 O}{\partial \sigma^2}(d\sigma)^2\} + \\ \{\frac{\partial^2 O}{\partial S \partial t}dSdt + \frac{\partial^2 O}{\partial S \partial \sigma}dS d\sigma + \frac{\partial^2 O}{\partial t\partial \sigma}dt d\sigma\} \\- \Delta dS - \sum_{i=1}^3 x_i \frac{\partial O}{\partial S} d S-\sum_{i=1}^3 x_i \frac{\partial C_i}{\partial t} d t-\sum_{i=1}^3 x_i \frac{\partial C_i}{\partial \sigma} d \sigma - \frac{1}{2}\left\{\sum_{i=1}^3 x_i \frac{\partial^2 C_i}{\partial S^2}\left(d S_t\right)^2+\sum_{i=1}^3 x_i \frac{\partial^2 C_i}{\partial t^2}(d t)^2+\sum_{i=1}^3 x_i \frac{\partial^2 C_i}{\partial \sigma^2}\left(d \sigma\right)^2\right\}  -\left\{\sum_{i=1}^3 x_i \frac{\partial^2 C_i}{\partial S \partial t} d S d t+\sum_{i=1}^3 x_i \frac{\partial^2 C_i}{\partial S \partial \sigma} d S d \sigma+\sum_{i=1}^3 x_i \frac{\partial^2 C_i}{\partial t \partial \sigma} d t d \sigma\right\}\tag{1.4}
$$

Since, 
$$(dS)^2 = \sigma^2 S^2 dt \tag{1.5}$$

By sorting out terms, we got:

$$ d \Pi^{B S}=\left[\frac{\partial O^{B S}}{\partial S_t}-\Delta_t-\sum_{i=1}^3 x_i \frac{\partial O^{B S}}{\partial S_t}\right] d S_t +\left[\left(\frac{\partial O^{B S}}{\partial t}-\sum_{i=1}^3 x_i \frac{\partial C_i^{B S}}{\partial t}\right)+\frac{1}{2} \sigma_t^2 S_t^2\left(\frac{\partial^2 O^{B S}}{\partial S_t^2}-\sum_{i=1}^3 x_i \frac{\partial^2 C_i^{B S}}{\partial S_t^2}\right)\right] d t +\left[\frac{\partial O^{B S}}{\partial \sigma_t}-\sum_{i=1}^3 x_i \frac{\partial C_i^{B S}}{\partial \sigma_t}\right] d \sigma_t+\frac{1}{2}\left[\partial^2 O^{B S} \partial \sigma_t^2-\sum_{i=1}^3 x_i \frac{\partial^2 C_i^{B S}}{\partial \sigma_t^2}\right]\left(d \sigma_t\right)^2  +\left[\frac{\partial^2 O^{B S}}{\partial S_t \partial \sigma_t}-\sum_{i=1}^3 x_i \frac{\partial^2 C_i^{B S}}{\partial S_t \partial \sigma_t}\right] d S_t d \sigma_t \tag{1.6}$$

Again, by Ito Lemma, 

$$d \Pi^{B S}=r_d \Pi^{B S} d t \tag{1.7}$$

$$ vega = \frac{\partial O}{\partial \sigma_t} \tag{1.8}$$

$$ vanna = \frac{\partial^2 O}{\partial \sigma_t  \partial S_t} \tag{1.9}$$

$$ volga = \frac{\partial^2 O}{\partial \sigma_t^2} \tag{1.10}$$

## 1.2 Calculate Delta

To construct a locally hedging portfolio we choose $\Delta_t$ and $\mathbf{x} = (x_1, x_2, x_3)^T$ in such way
that they zero out the coefficients of $d{S_t}, d\sigma_t, d(\sigma_t)^2$ and $dS_td\sigma_t $, so that we get rid of the risk associated with the fluctuations of the spot price and volatility. It only matters with $t$.

The last three partial derivatives in equation (1.6), is vega, volga and vanna.

Since, 
$$ \frac{\partial O^{B S}}{\partial S_t}-\Delta_t-\sum_{i=1}^3 x_i \frac{\partial O^{B S}}{\partial S_t}= 0 \tag{1.11}
$$

Then,

$$
\Delta_t=\frac{\partial O^{B S}}{\partial S_t}-\sum_{i=1}^3 x_i \frac{\partial O^{B S}}{\partial S_t} \tag{1.12}$$

$$\mathbf{w}_{\mathbf{O}}=\mathbf{V} \mathbf{x} \tag{1.13}$$

where

$$
\mathbf{w}_{\mathbf{O}}=\left(\begin{array}{c}
\operatorname{Vega}(O) \\
\operatorname{Vanna}(O) \\
\operatorname{Volga}(O)
\end{array}\right)=\left(\begin{array}{c}
\frac{\partial O^{B S}}{\partial \sigma_t} \\
\frac{\partial^2 O^{B S}}{\partial S_t \partial \sigma_t} \\
\frac{\partial^2 O^{B S}}{\partial \sigma_t^2}
\end{array}\right) \tag{1.14}
$$

$$
\mathbf{V}=\left(\begin{array}{ccc}
\text Vega_1 & \text Vega_2 & \text Vega_3 \\
\text Vanna_1 & \text Vanna_2 & \text Vanna_3 \\
\text Volga_1 & \text Volga_2 & \text Volga_3
\end{array}\right)=\left(\begin{array}{ccc}
\frac{\partial C_1^{B S}}{\partial \sigma_t} & \frac{\partial C_2^{B S}}{\partial \sigma_t} & \frac{\partial C_3^{B S}}{\partial \sigma_t} \\
\frac{\partial^2 C_1^{B S}}{\partial S_t \partial \sigma_t} & \frac{\partial^2 C_2^{B S}}{\partial S_t \partial \sigma_t} & \frac{\partial^2 C_3^{B S}}{\partial S_t \partial \sigma_t} \\
\frac{\partial^2 C_1^{B S}}{\partial \sigma_t^2} & \frac{\partial^2 C_2^{B S}}{\partial \sigma_t^2} & \frac{\partial^2 C_3^{B S}}{\partial \sigma_t^2}
\end{array}\right) \tag{1.15}$$

Assume $T_1 = T_2 = T_3$,

we got：

$$x_1=\frac{V e g a(O) \ln \frac{K_2}{K} \ln \frac{K_3}{K}}{V e g a_1 \ln \frac{K_2}{K_1} \ln \frac{K_3}{K_1}}\tag{1.16}$$

$$x_2=\frac{V e g a(O) \ln \frac{K}{K_1} \ln \frac{K_3}{K}}{V e g a_2 \ln \frac{K_2}{K_1} \ln \frac{K_3}{K_2}}\tag{1.17}$$

$$x_3=\frac{V e g a(O) \ln \frac{K}{K_1} \ln \frac{K}{K_2}}{V e g a_3 \ln \frac{K_3}{K_1} \ln \frac{K_3}{K_2}}\tag{1.18}$$

# 2. Vanna - Volga Pricing

## 2.1 The Exact Method

Since
$$ \Pi^{MK} = O^{MK} - \Delta_t^{MK}S_t - \sum_{i = 1}^3 x_idC_i^{MK} \tag{2.1}$$
and
$$ \Pi^{MK} = \Pi^{BS} \tag{2.2}$$ 
for any $0 \leq t \leq T $, it follows that:

$$O_{V V}^{M K}=O^{B S}+\sum_{i=1}^3 x_i\left(C_i^{M K}-C_i^{B S}\right) \tag{2.3}$$

where the last term is called **overhedge**.

Butterfly Spread:


Risk Reversal:


$$\operatorname{Risk} \operatorname{Reversal}\left(K_C, K_P\right)=\operatorname{Call}\left(K_C, \sigma\left(K_C\right)\right)-\operatorname{Put}\left(K_P, \sigma\left(K_P\right)\right) \tag{2.4}$$

$$
\operatorname{Strangle}\left(K_C, K_P\right)=\operatorname{Call}\left(K_C, \sigma\left(K_C\right)\right)+\operatorname{Put}\left(K_P, \sigma\left(K_P\right)\right), \\
\operatorname{Straddle}(K)=\operatorname{Call}\left(K, \sigma_{A T M}\right)+\operatorname{Put}\left(K, \sigma_{A T M}\right)\tag{2.5}
$$
$$ ATM(K) = \frac{1}{2}Straddle(K) \tag{2.6}$$
$$Butterfly \left(K_P, K, K_C\right)=\frac{1}{2}\left[\operatorname{Strangle}\left(K_C, K_P\right)-\operatorname{Straddle}(K)\right] \tag{2.7}$$


For volatility,
$$\sigma_{RR25} = \sigma_{25\Delta C} - \sigma_{25\Delta P} \tag{2.8}$$
$$\sigma_{BF25} = \frac{1}{2}[\sigma_{25\Delta C} + \sigma_{25\Delta P}] - \sigma_{ATM}\tag{2.9}$$

Therefore, using eq(2.8) and (2.9), we can use BF, ATM and RR to represent C and P
$$ C = BF + \frac{1}{2}RR + ATM \tag{2.10}$$
$$ P = BF - \frac{1}{2}RR + ATM \tag{2.11}$$

We can denote x in another way,

$$\tilde{x}_{A T M}=x_1+x_2+x_3 \tag{2.12}$$
$$\tilde{x}_{R R}=\frac{1}{2}\left(x_3-x_1\right)\tag{2.12}$$ 
$$\tilde{x}_{B F}=x_1+x_3 \tag{2.13}$$

so that our matrix becomes:

$$\mathbf{w}_{\mathbf{O}}=\mathbf{A} \mathbf {\tilde{x}} \tag{2.14}$$

$$\mathbf{w}_{\mathbf{O}}=\left(\begin{array}{c}
\operatorname{Vega}(O) \\
\operatorname{Vanna}(O) \\
\operatorname{Volga}(O)
\end{array}\right) \tag{2.15}$$

$${\mathbf{A}}=\left(\begin{array}{c}
\operatorname{Vega}(ATM) & \operatorname{Vega}(RR) & \operatorname{Vega}(BF)\\
\operatorname{Vanna}(ATM)& \operatorname{Vanna}(RR) & \operatorname{Vanna}(BF)\\
\operatorname{Volga}(ATM) & \operatorname{Volga}(RR) & \operatorname{Volga}(BF)
\end{array}\right) \tag{2.16}$$

$${\mathbf{\tilde{x}}}=\left(\begin{array}{c}
\tilde{x}_{ATM}\\
\tilde{x}_{RR} \\
\tilde{x}_{BF}
\end{array}\right) \tag{2.17}$$

And the VV formula is :

$$O_{V V}^{M K}=O^{B S}+ \tilde{x}^T\mathbf{Y}\tag{2.17}$$

where,
$${\mathbf{Y}}=\left(\begin{array}{c}
ATM_{MK} - ATM_{BS}\\
RR_{MK} - RR_{BS} \\
BF_{MK} - BF_{BS}
\end{array}\right) \tag{2.18}$$

In other words,

$$O_{V V}^{M K}=O^{B S}+\tilde{x}_{A T M}\left(A T M^{M K}-A T M^{B S}\right)+\tilde{x}_{R R}\left(R R^{M K}-R R^{B S}\right)+\tilde{x}_{B F}\left(B F^{M K}-B F^{B S}\right) \tag{2.19}$$

Or (exactly the same):

$$O_{V V}^{M K}=O^{B S}+w_O^T\left(A^T\right)^{-1} Y=O^{B S}+V e g a(O) \Omega_{v e g a}+V a n n a(O) \Omega_{v a n n a}+V o l g a(O) \Omega_{v o l g a} \tag{2.20}$$

where
$$\omega =\left(\begin{array}{c}
\Omega_{vega}\\
\Omega_{vanna} \\
\Omega_{volga}
\end{array}\right) \tag{2.21}$$

Moreover, we can retrive from delta that:

$$ K = S0 e^{(r_d - r_f)\tau +  N^{-1} (\Delta) \sigma \sqrt\tau + [(x_d - x_f) + 0.5 \sigma^2]  \tau} \tag{2.22}$$



By expanding both members of eq(2.3) at first order in $\sigma = \sigma_2$,

1st Order Implied Vol can be approximated by:

$$ \sigma' = x_1 \sigma_1 + x_2\sigma_2 + x_3 \sigma_3  \tag{2.23}$$

which is extremely accurate inside the interval $(K_1, K_3)$. However, it tends to be overcalued. In fact, being the quadratic functional form in the log-strike, the no-arbitrage conditions derived by Lee (2004) for the asymptotic value of implied volatilities are violated. 

We use 2nd Order Implied Vol to address this problem:

$$ \sigma'' = \sigma_2^2 + \frac{\sqrt{-\sigma_2 + d_1(K)d_2(K)(2\sigma_2 D_1(K)D_2(K))}}{d_1(K)d_2(K)}  \tag{2.24}$$

where

$$ D_1(K) = \sigma' - \sigma_2  \tag{2.25}$$

$$ D_2(K) = x_1 d_1(K_1)d_2(K_1)(\sigma_1 - \sigma_2)^2 + x_3 d_3(K_3)d_2(K_3)(\sigma_3 - \sigma_2)^2  \tag{2.26}$$

## 2.2 The Simplified Method

$$O_{V V}^{M K}=O^{B S}+\frac{\operatorname{Vanna}(O)}{\operatorname{Vanna}(R R)} R R_{\text {cost }}+\frac{\operatorname{Volga}(O)}{\operatorname{Volga}(B F)} B F_{\text {cost }}\tag{2.23}$$

where,

$$R R_{\text {cost }}= \left[\operatorname{Call}\left(K_c, \sigma\left(K_c\right)\right)-\operatorname{Put}\left(K_p, \sigma\left(K_p\right)\right)\right] -\left[\operatorname{Call}\left(K_c, \sigma_{A T M}\right)-\operatorname{Put}\left(K_p, \sigma_{A T M}\right)\right] \tag{2.24}$$

$$B F_{\text {cost }}= \frac{1}{2}\left[\operatorname{Call}\left(K_c, \sigma\left(K_c\right)\right)+\operatorname{Put}\left(K_p, \sigma\left(K_p\right)\right)\right] -\frac{1}{2}\left[\operatorname{Call}\left(K_c, \sigma_{A T M}\right)+\operatorname{Put}\left(K_p, \sigma_{A T M}\right)\right] \tag{2.25}$$

## 2.3 Price the First-Gen Exotics

Now，we have the formula for vanilla, but it doesn't make sense for barriers.

So, we involve $p \in [0, 1]$，the survival probability. The exact values of p，will be determined by traders, because for different products, different markets, and different traders, they might have different opinions.

$$O_{V V}^{M K}=O^{B S}+p\sum_{i=1}^3 x_i\left(C_i^{M K}-C_i^{B S}\right) \tag{2.26}$$

$$O_{V V}^{M K}=O^{B S}+ p[V e g a(O) \Omega_{v e g a}+V a n n a(O) \Omega_{v a n n a}+V o l g a(O) \Omega_{v o l g a}] \tag{2.27}$$

$$O_{V V}^{M K}=O^{B S}+p[\frac{\operatorname{Vanna}(O)}{\operatorname{Vanna}(R R)} R R_{\text {cost }}+\frac{\operatorname{Volga}(O)}{\operatorname{Volga}(B F)} B F_{\text {cost }}] \tag{2.28}$$


where
$$p = 1 - q = 1 - P[\tau_B \leq T] \tag{2.29}$$
q is the ko probability，$\tau_B$ is the first hitting time

Therefore,
$$knock out + knock in = vanilla \tag{2.30}$$

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from scipy import interpolate
import scipy.stats as sci
from scipy.stats import norm

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

from tqdm import tqdm

import warnings
warnings.filterwarnings('ignore')
```

```python
class Vanna_Volga:
    """Vanna-Volga Implied Volatility method used to price an FX option"""

    def __init__(self, S0, strike, t, importfile, num_interp):
        self.dc = 360
        self.S0 = S0
        self.strike = strike
        self.t = t
        self.importfile = importfile
        self.num_interp = num_interp

    def rate(self, xf, xd, vol_data):
        """Interpolate xd and xf, and then shape them as vol_data"""
        xd_array = np.array(xd).squeeze()
        xf_array = np.array(xf).squeeze()
        
        T = pd.DataFrame({'X': vol_data['Term'].unique()})
        xd_interp = interpolate.interp1d(np.arange(xd_array.size), xd_array, kind='cubic')
        xd_compress = xd_interp(np.linspace(0, T.size - 1, self.num_interp))
        
        xf_interp = interpolate.interp1d(np.arange(xf_array.size), xf_array, kind='cubic')
        xf_compress = xf_interp(np.linspace(0, T.size - 1, self.num_interp))
    
        fig = plt.figure(figsize = (12, 3))
        plt.plot(np.linspace(0, T.size - 1, 1000), xd_interp(np.linspace(0, xd_array.size - 1, 1000)), color = "blue")
        plt.plot(np.linspace(0, T.size - 1, 1000), xf_interp(np.linspace(0, xf_array.size - 1, 1000)), color = "red")
        
        plt.title('US and EU Deposit Rate Intepolation', fontsize = 10)
        plt.legend(['US Deposit Rate', 'EU Deposit Rate'], loc = 0)
        plt.xticks(np.linspace(0, T.size - 1, len(T)), labels=T['X'], rotation=45)
        plt.show()
        return xd_compress, xf_compress

    def delta25_strike (self, T, sig25DeltaPut, sigATM, sig25DeltaCall, xd, xf):
        """Calculate the strikes K25DeltaPut, KATM, K25DeltaCall"""
        print
        Dd = np.exp(-np.multiply(xd, T))
        Df = np.exp(-np.multiply(xf, T))
        delta = 0.25
        alpha = -sci.norm.ppf(delta * np.reciprocal(Df))
        mu = xd - xf
        F = self.S0 * np.exp(np.multiply(mu, T))

        KATM = np.array([])
        K25DeltaCall = np.array([])
        K25DeltaPut = np.array([])

        for i in range(len(T)):
            K25DeltaPut = np.append(K25DeltaPut, F[i] * math.exp(
                -(alpha[i] * sig25DeltaPut[i] * math.sqrt(T[i])) + (xd[i] - xf[i] + 0.5 * sig25DeltaPut[i] ** 2) * T[i]))
            KATM = np.append(KATM, F[i] * math.exp(xd[i] - xf[i] + 0.5 * T[i] * (sigATM[i]) ** 2))
            K25DeltaCall = np.append(K25DeltaCall, F[i] * math.exp(
                alpha[i] * sig25DeltaCall[i] * math.sqrt(T[i]) + (xd[i] - xf[i] + 0.5 * sig25DeltaCall[i] ** 2) * T[i]))

        return K25DeltaPut, KATM, K25DeltaCall, F

    def d1(self, F, K, sig, T, xd, xf):
        return (np.log(F / K) + (xd - xf + 0.5 * (sig ** 2)) * T) / (sig * np.sqrt(T))

    def d2(self, F, K, sig, T, xd, xf):
        return self.d1(F, K, sig, T, xd, xf) - sig * np.sqrt(T)

    def bs_price(self, F, K, T, xd, xf, sig):
        bs_val = norm.cdf(self.d1(F, K, sig, T, xd, xf)) * F * math.exp(-xf * T) - norm.cdf(
            self.d2(F, K, sig, T, xd, xf)) * F * math.exp(-xd * T)

        return bs_val
    
    def delta(self, F, K, T, xd, xf, sig, j):
        return math.exp(-xf * T) * norm.cdf(self.d1(F, K, sig, T, xd, xf))
    
    def gamma(self, F, K, T, xd, xf, sig, j):
        d1_val = self.d1(F, K, sig, T, xd, xf)
        gamma_val = (norm.pdf(d1_val) * np.exp(-xf * T)) / (F * sig * np.sqrt(T))
        return gamma_val
    
    def theta(self, F, K, T, xd, xf, sig, j):
        d1_val = self.d1(F, K, sig, T, xd, xf)
        d2_val = self.d2(F, K, sig, T, xd, xf)
        theta_val = (-F * norm.pdf(d1_val) * sig * np.exp(-xf * T)) / (2 * np.sqrt(T)) - \
                    (xd - xf) * F * norm.cdf(d1_val) * np.exp(-xf * T)
        return theta_val
    
    def rho(self, F, K, T, xd, xf, sig, j):
        d2_val = self.d2(F, K, sig, T, xd, xf)
        rho_val = K * T * np.exp(-xd * T) * norm.cdf(d2_val)
        return rho_val
    
    def vega(self, F, K, T, xd, xf, sig):
        d1_val = self.d1(F, K, sig, T, xd, xf)
        vega_val = F * np.exp(-xf * T) * np.sqrt(T) * norm.pdf(d1_val)
        return vega_val
    
    def vega_new(self, F, K, T, xd, xf, sig, j):
        d1_val = self.d1(F, K, sig, T, xd, xf)
        vega_val = F * np.exp(-xf * T) * np.sqrt(T) * norm.pdf(d1_val)
        return vega_val
    
    def vanna(self, F, K, T, xd, xf, sig, j):
        d1_val = self.d1(F, K, sig, T, xd, xf)
        d2_val = self.d2(F, K, sig, T, xd, xf)
        vanna_val = -norm.pdf(d1_val) * d2_val / sig
        return vanna_val
    
    def volga(self, F, K, T, xd, xf, sig, j):
        d1_val = self.d1(F, K, sig, T, xd, xf)
        d2_val = self.d2(F, K, sig, T, xd, xf)
        volga_val = F * np.exp(-xf * T) * np.sqrt(T) * norm.pdf(d1_val) * d1_val * d2_val / sig
        return volga_val
    
    def implied_price(self, F, K, T, K1, K2, K3, sig1, sig2, sig3, xd, xf, j):
        v_v1 = self.vega(F, K, T, xd, xf, sig2) / self.vega(F, K1, T, xd, xf, sig1)
        v_v2 = self.vega(F, K, T, xd, xf, sig2) / self.vega(F, K2, T, xd, xf, sig2)
        v_v3 = self.vega(F, K, T, xd, xf, sig2) / self.vega(F, K3, T, xd, xf, sig3)

        x1 = v_v1 * ((math.log(K2 / K) * math.log(K3 / K)) / (math.log(K2 / K1) * math.log(K3 / K1)))
        x2 = v_v2 * ((math.log(K / K1) * math.log(K3 / K)) / (math.log(K2 / K1) * math.log(K3 / K2)))
        x3 = v_v3 * ((math.log(K / K1) * math.log(K / K2)) / (math.log(K3 / K1) * math.log(K3 / K2)))

        a = self.bs_price(F, K2, T, xd, xf, sig2)

        vv_price = self.bs_price(F, K2, T, xd, xf, sig2) + x1 * (
        self.bs_price(F, K1, T, xd, xf, sig1) - self.bs_price(F, K1, T, xd, xf, sig2))
        + x3 * (self.bs_price(F, K3, T, xd, xf, sig3) - self.bs_price(F, K3, T, xd, xf, sig2))

        return vv_price
    
    
    def implied_vol_vv(self, F, K, T, K1, K2, K3, sig1, sig2, sig3, xd, xf, j):
        x1 = (math.log(K2 / K) * math.log(K3 / K)) / (math.log(K2 / K1) * math.log(K3 / K1))
        x2 = (math.log(K / K1) * math.log(K3 / K)) / (math.log(K2 / K1) * math.log(K3 / K2))
        x3 = (math.log(K / K1) * math.log(K / K2)) / (math.log(K3 / K1) * math.log(K3 / K2))
        
        # Get the 1st order implied vol
        D1 = (x1 * sig1 + x2 * sig2 + x3 * sig3) - sig2

        D2 = (x1 * self.d1(F, K1, sig1, T, xd, xf) * self.d2(F, K1, sig1, T, xd, xf) * (sig1 - sig2) ** 2) \
             + (x3 * self.d1(F, K3, sig3, T, xd, xf) * self.d2(F, K3, sig3, T, xd, xf) * (sig3 - sig2) ** 2)
        
        # Get the 2nd order implied vol
        sig = sig2 + (-sig2 + np.sqrt(sig2 ** 2 + self.d1(F, K, sig2, T, xd, xf) * self.d2(F, K, sig2, T, xd, xf) * (2 * sig2 * D1 + D2))) \
                     / (self.d1(F, K, sig2, T, xd, xf) * self.d2(F, K, sig2, T, xd, xf))

        return sig

    def create_3dplots(self, F, T, K25DeltaPut, KATM, K25DeltaCall, sig25DeltaPut, sigATM, sig25DeltaCall, xd, xf):
        m = len(T[:-1])

        K_array = np.linspace(0.8 * self.S0, 1.4 * self.S0, m)

        # Vol, Price & Greeks
        implied_matrix = np.zeros((m, m), dtype=float)
        price_matrix = np.zeros((m, m), dtype=float)
        delta_matrix = np.zeros((m, m), dtype=float)
        gamma_matrix = np.zeros((m, m), dtype=float)
        theta_matrix = np.zeros((m, m), dtype=float)
        vega_matrix = np.zeros((m, m), dtype=float)
        rho_matrix = np.zeros((m, m), dtype=float)
        vanna_matrix = np.zeros((m, m), dtype=float)
        volga_matrix = np.zeros((m, m), dtype=float)

        for i in tqdm(range(K_array.size)):
            for j in range(m):
                implied_matrix[i][j] = self.implied_vol_vv(F[j], K_array[i], T[j], K25DeltaPut[j], KATM[j],
                                                        K25DeltaCall[j], sig25DeltaPut[j], sigATM[j], sig25DeltaCall[j], xd[j], xf[j], j)
                price_matrix[i][j] = self.implied_price(F[j], K_array[i], T[j], K25DeltaPut[j], KATM[j],
                                                        K25DeltaCall[j], sig25DeltaPut[j], sigATM[j], sig25DeltaCall[j], xd[j], xf[j], j)
                delta_matrix[i][j] = self.delta(F[j], K_array[i], T[j], xd[j], xf[j], implied_matrix[i][j], j)
                gamma_matrix[i][j] = self.gamma(F[j], K_array[i], T[j], xd[j], xf[j], implied_matrix[i][j], j)
                theta_matrix[i][j] = self.theta(F[j], K_array[i], T[j], xd[j], xf[j], implied_matrix[i][j], j)
                vega_matrix[i][j] = self.vega_new(F[j], K_array[i], T[j], xd[j], xf[j], implied_matrix[i][j], j)
                rho_matrix[i][j] = self.rho(F[j], K_array[i], T[j], xd[j], xf[j], implied_matrix[i][j], j)
                vanna_matrix[i][j] = self.vanna(F[j], K_array[i], T[j], xd[j], xf[j], implied_matrix[i][j], j)
                volga_matrix[i][j] = self.volga(F[j], K_array[i], T[j], xd[j], xf[j], implied_matrix[i][j], j)

        implied_matrix = pd.DataFrame(implied_matrix)
        price_matrix = pd.DataFrame(price_matrix)
        delta_matrix = pd.DataFrame(delta_matrix)
        gamma_matrix = pd.DataFrame(gamma_matrix)
        theta_matrix = pd.DataFrame(theta_matrix)
        vega_matrix = pd.DataFrame(vega_matrix)
        rho_matrix = pd.DataFrame(rho_matrix)
        vanna_matrix = pd.DataFrame(vanna_matrix)
        volga_matrix = pd.DataFrame(volga_matrix)
        
        implied_matrix.fillna(method='ffill', axis=1, inplace=True)
        price_matrix.fillna(method='ffill', axis=1, inplace=True)
        delta_matrix.fillna(method='ffill', axis=1, inplace=True)
        gamma_matrix.fillna(method='ffill', axis=1, inplace=True)
        theta_matrix.fillna(method='ffill', axis=1, inplace=True)
        vega_matrix.fillna(method='ffill', axis=1, inplace=True)
        rho_matrix.fillna(method='ffill', axis=1, inplace=True)
        vanna_matrix.fillna(method='ffill', axis=1, inplace=True)
        volga_matrix.fillna(method='ffill', axis=1, inplace=True)
        
        #  Flatten the matrices for use with griddata
        implied_flat = implied_matrix.values.ravel()
        price_flat = price_matrix.values.ravel()
        delta_flat = delta_matrix.values.ravel()
        gamma_flat = gamma_matrix.values.ravel()
        theta_flat = theta_matrix.values.ravel()
        vega_flat = vega_matrix.values.ravel()
        rho_flat = rho_matrix.values.ravel()
        vanna_flat = vanna_matrix.values.ravel()
        volga_flat = volga_matrix.values.ravel()

        KK, TT = np.meshgrid(np.array(K_array), np.array(T[:-1]), indexing='ij')

        option_vol = interpolate.griddata((KK.ravel(), TT.ravel()), implied_flat, ([self.t], [self.strike]), method='nearest')
        option_premium = interpolate.griddata((KK.ravel(), TT.ravel()), price_flat, ([self.t], [self.strike]), method='nearest')
        option_delta = interpolate.griddata((KK.ravel(), TT.ravel()), delta_flat, ([self.t], [self.strike]), method='nearest')
        option_gamma = interpolate.griddata((KK.ravel(), TT.ravel()), gamma_flat, ([self.t], [self.strike]), method='nearest')
        option_theta = interpolate.griddata((KK.ravel(), TT.ravel()), theta_flat, ([self.t], [self.strike]), method='nearest')
        option_vega = interpolate.griddata((KK.ravel(), TT.ravel()), vega_flat, ([self.t], [self.strike]), method='nearest')
        option_rho = interpolate.griddata((KK.ravel(), TT.ravel()), rho_flat, ([self.t], [self.strike]), method='nearest')
        option_vanna = interpolate.griddata((KK.ravel(), TT.ravel()), vanna_flat, ([self.t], [self.strike]), method='nearest')
        option_volga = interpolate.griddata((KK.ravel(), TT.ravel()), volga_flat, ([self.t], [self.strike]), method='nearest')
        
        
        #  Gridspec
        gs = gridspec.GridSpec(10, 10)

        #  Gridspec
        gs = gridspec.GridSpec(3, 2)

        #  3D Meshgrid of T, K and Volatility
        X = T[:-1]
        Y = K_array
        X, Y = np.meshgrid(X, Y)
        Z = 100*implied_matrix
        fig = plt.figure()
        ax1 = fig.add_subplot(gs[:-1, :], projection='3d')
        ax1.set_xlabel('Tenor (yrs)')
        ax1.set_ylabel('Strike')
        ax1.set_zlabel('Implied Volatility (%)')
        plt.suptitle('Option Greeks')
        surf = ax1.plot_surface(X, Y, Z, cmap='turbo', linewidth=5)
        fig.set_size_inches(13, 10)
        
        #  Implied Volatility vs Strikes at 3M
        X = K_array
        Y = 100*implied_matrix[:][self.num_interp - 2]
        ax2 = fig.add_subplot(gs[-1, 0])
        ax2.set_xlabel('Strike')
        ax2.set_ylabel('Implied Volatility (%)')
        plt.title('Implied Volatility vs Strikes at 10Y')
        f = ax2.plot(X, Y, color = 'blue')

        #  Option Premium vs Strikes at 3M
        X = K_array
        Y = price_matrix[:][self.num_interp - 2]
        ax3 = fig.add_subplot(gs[-1, -1])
        ax3.set_xlabel('Strike')
        ax3.set_ylabel('C(K) (USD)')
        plt.title('Option Premium vs Strikes at 10Y')
        g = ax3.plot(X, Y, color = 'blue')
        
        plt.show()
        return 'Option Volatility: {:.4f}%\nOption Premium: {:.4f}%\nDelta: {:.4f}\nGamma: {:.4f}\nTheta: {:.4f}\nVega: {:.4f}\nRho: {:.4f}\nVanna: {:.4f}\nVolga: {:.4f}\n'.\
    format(option_vol[0] * 100, option_premium[0], option_delta[0], option_gamma[0], option_theta[0], option_vega[0], option_rho[0], option_vanna[0], option_volga[0])

    
    def FXOptionPrice(self):

        xlsx_file = pd.ExcelFile(self.importfile)
        vol_data = xlsx_file.parse('Volatility')

        # US Deposit Data
        xd_data = xlsx_file.parse('US_Depo')
        xd = np.array(xd_data['Value'])
        T_xd = pd.DataFrame({'X': xd_data['Term']})

        # EU Deposit Data
        xf_data = xlsx_file.parse('EU_Depo')
        xf = np.array(xf_data['Value'])
        T_xf = pd.DataFrame({'X': xf_data['Term']})

        T = pd.DataFrame({'X': vol_data['Term'].unique()})  # terms

        # Convert to day/360 format
        super_dict = [('M', '*30'), ('D', '*1'), ('W', '*7'), ('Y', '*360'), (' ', '')]
        super_dict = dict(super_dict)

        # Modify the terms to daycount and store the vols in numpy arrays
        T = T.replace(super_dict, regex=True)
        T_orig = T.replace(super_dict, regex=True)
        T_xd = T_xd.replace(super_dict, regex=True)
        T_xf = T_xf.replace(super_dict, regex=True)

        T = T.X.apply(lambda x: eval(str(x))) / self.dc
        T_xd = T_xd.X.apply(lambda x: eval(str(x))) / self.dc
        T_xf = T_xf.X.apply(lambda x: eval(str(x))) / self.dc
        
        # T interp
        T_interp = interpolate.interp1d(np.arange(T.size), T, kind='cubic')
        T_compress = T_interp(np.linspace(0, T.size - 1, self.num_interp))
        T = T_compress

        sigATM_orig = np.array(vol_data[vol_data['Strike'] == 'ATM']['Value']) / 100
        sigATM = interpolate.interp1d(np.arange(T_orig.size), sigATM_orig, kind = 'cubic')
        sigATM = sigATM(np.linspace(0, T_orig.size - 1, self.num_interp))
        
        sig25DeltaRR_orig = np.array(vol_data[vol_data['Strike'] == '25 Delta RR']['Value']) / 100
        sig25DeltaRR = interpolate.interp1d(np.arange(T_orig.size), sig25DeltaRR_orig, kind = 'cubic')
        sig25DeltaRR = sig25DeltaRR(np.linspace(0, T_orig.size - 1, self.num_interp))
        
        sig25DeltaBF = np.array(vol_data[vol_data['Strike'] == '25 Delta BF']['Value']) / 100
        sig25DeltaBF = interpolate.interp1d(np.arange(T_orig.size), sig25DeltaBF, kind = 'cubic')
        sig25DeltaBF = sig25DeltaBF(np.linspace(0, T_orig.size - 1, self.num_interp))        

        sig25DeltaCall = sig25DeltaBF + sigATM + sig25DeltaRR / 2
        sig25DeltaPut = sig25DeltaBF + sigATM - sig25DeltaRR / 2
        
        xd, xf = self.rate(xf, xd, vol_data)
        K25DeltaPut, KATM, K25DeltaCall, F = self.delta25_strike(T, sig25DeltaPut, sigATM, sig25DeltaCall, xd, xf)
        
        option_data = self.create_3dplots(F, T, K25DeltaPut, KATM, K25DeltaCall, sig25DeltaPut, sigATM, sig25DeltaCall, xd, xf)

        return option_data

def main():

    validus_answer = Vanna_Volga(1.0977, 1.1, 10, 'vanna-volga.xlsx', 50).FXOptionPrice()
    print(validus_answer)
    plt.show()

if __name__ == "__main__":
    main()
```

```python
class Vanna_Volga:
    """Vanna-Volga Implied Volatility method used to price an FX option"""

    def __init__(self, S0, strike, t, importfile):
        self.dc = 360
        self.S0 = S0
        self.strike = strike
        self.t = t
        self.importfile = importfile

    def rate(self, xf, xd, vol_data):
        """Interpolate xd and xf, and then shape them as vol_data"""
        xd_array = np.array(xd).squeeze()
        xf_array = np.array(xf).squeeze()
        
        T = pd.DataFrame({'X': vol_data['Term'].unique()})
        xd_interp = interpolate.interp1d(np.arange(xd_array.size), xd_array, kind='cubic')
        xd_compress = xd_interp(np.linspace(0, T.size - 1, T.size))
        
        xf_interp = interpolate.interp1d(np.arange(xf_array.size), xf_array, kind='cubic')
        xf_compress = xf_interp(np.linspace(0, T.size - 1, T.size))

        return xd_compress, xf_compress

    def delta25_strike (self, T, sig25DeltaPut, sigATM, sig25DeltaCall, xd, xf):
        """Calculate the strikes K25DeltaPut, KATM, K25DeltaCall"""
        Dd = np.exp(-np.multiply(xd, T))
        Df = np.exp(-np.multiply(xf, T))
        delta = 0.25
        alpha = -sci.norm.ppf(delta * np.reciprocal(Df))
        mu = xd - xf
        F = self.S0 * np.exp(np.multiply(mu, T))

        KATM = np.array([])
        K25DeltaCall = np.array([])
        K25DeltaPut = np.array([])

        for i in range(len(T)):
            K25DeltaPut = np.append(K25DeltaPut, F[i] * math.exp(
                -(alpha[i] * sig25DeltaPut[i] * math.sqrt(T[i])) + (xd[i] - xf[i] + 0.5 * sig25DeltaPut[i] ** 2) * T[i]))
            KATM = np.append(KATM, F[i] * math.exp(xd[i] - xf[i] + 0.5 * T[i] * (sigATM[i]) ** 2))
            K25DeltaCall = np.append(K25DeltaCall, F[i] * math.exp(
                alpha[i] * sig25DeltaCall[i] * math.sqrt(T[i]) + (xd[i] - xf[i] + 0.5 * sig25DeltaCall[i] ** 2) * T[i]))

        return K25DeltaPut, KATM, K25DeltaCall, F

    def d1(self, F, K, sig, T, xd, xf):
        return (np.log(F / K) + (xd - xf + 0.5 * (sig ** 2)) * T) / (sig * np.sqrt(T))

    def d2(self, F, K, sig, T, xd, xf):
        return self.d1(F, K, sig, T, xd, xf) - sig * np.sqrt(T)

    def bs_price(self, F, K, T, xd, xf, sig):
        bs_val = norm.cdf(self.d1(F, K, sig, T, xd, xf)) * F * math.exp(-xf * T) - norm.cdf(
            self.d2(F, K, sig, T, xd, xf)) * F * math.exp(-xd * T)

        return bs_val
    
    def vega(self, F, K, T, xd, xf, sig):
        d1_val = self.d1(F, K, sig, T, xd, xf)
        vega_val = F * np.exp(-xf * T) * np.sqrt(T) * norm.pdf(d1_val)
        return vega_val

    def delta(self, F, K, T, xd, xf, sig, j):
        return math.exp(-xf * T) * norm.cdf(self.d1(F, K, sig, T, xd, xf))
    
    def implied_price(self, F, K, T, K1, K2, K3, sig1, sig2, sig3, xd, xf, j):
        v_v1 = self.vega(F, K, T, xd, xf, sig2) / self.vega(F, K1, T, xd, xf, sig1)
        v_v2 = self.vega(F, K, T, xd, xf, sig2) / self.vega(F, K2, T, xd, xf, sig2)
        v_v3 = self.vega(F, K, T, xd, xf, sig2) / self.vega(F, K3, T, xd, xf, sig3)

        x1 = v_v1 * ((math.log(K2 / K) * math.log(K3 / K)) / (math.log(K2 / K1) * math.log(K3 / K1)))
        x2 = v_v2 * ((math.log(K / K1) * math.log(K3 / K)) / (math.log(K2 / K1) * math.log(K3 / K2)))
        x3 = v_v3 * ((math.log(K / K1) * math.log(K / K2)) / (math.log(K3 / K1) * math.log(K3 / K2)))

        a = self.bs_price(F, K2, T, xd, xf, sig2)

        vv_price = self.bs_price(F, K2, T, xd, xf, sig2) + x1 * (
        self.bs_price(F, K1, T, xd, xf, sig1) - self.bs_price(F, K1, T, xd, xf, sig2))
        + x3 * (self.bs_price(F, K3, T, xd, xf, sig3) - self.bs_price(F, K3, T, xd, xf, sig2))

        return vv_price
    
    
    def implied_vol_vv(self, F, K, T, K1, K2, K3, sig1, sig2, sig3, xd, xf, j):
        x1 = (math.log(K2 / K) * math.log(K3 / K)) / (math.log(K2 / K1) * math.log(K3 / K1))
        x2 = (math.log(K / K1) * math.log(K3 / K)) / (math.log(K2 / K1) * math.log(K3 / K2))
        x3 = (math.log(K / K1) * math.log(K / K2)) / (math.log(K3 / K1) * math.log(K3 / K2))
        
        # get the 1st order implied vol
        D1 = (x1 * sig1 + x2 * sig2 + x3 * sig3) - sig2

        D2 = (x1 * self.d1(F, K1, sig1, T, xd, xf) * self.d2(F, K1, sig1, T, xd, xf) * (sig1 - sig2) ** 2) \
             + (x3 * self.d1(F, K3, sig3, T, xd, xf) * self.d2(F, K3, sig3, T, xd, xf) * (sig3 - sig2) ** 2)
        
        # get the 2nd order implied vol
        sig = sig2 + (-sig2 + np.sqrt(sig2 ** 2 + self.d1(F, K, sig2, T, xd, xf) * self.d2(F, K, sig2, T, xd, xf) * (2 * sig2 * D1 + D2))) \
                     / (self.d1(F, K, sig2, T, xd, xf) * self.d2(F, K, sig2, T, xd, xf))

        return sig
    
    def interpolate_volatility(self, terms, strikes, vol_matrix, new_terms, new_strikes):
        interp_vol_matrix = np.zeros((len(new_terms), len(new_strikes)))

        for i, new_term in enumerate(new_terms):
            term_indices = np.where(terms == new_term)[0]
            for j, new_strike in enumerate(new_strikes):
                strike_indices = np.where(strikes == new_strike)[0]

                # Interpolate using the available data points
                interp_vol = interpolate.interp2d(strike_indices, term_indices, vol_matrix[term_indices, :][:, strike_indices], kind='linear')
                interp_vol_matrix[i, j] = interp_vol(0, 0)  # Interpolate at the first available point

        return interp_vol_matrix


    def create_plots(self, F, T, K25DeltaPut, KATM, K25DeltaCall, sig25DeltaPut, sigATM, sig25DeltaCall, xd, xf):
        m = len(T[:-1])

        K_array = np.linspace(0.8 * self.S0, 1.4 * self.S0, m)

        # vol, price & delta
        implied_matrix = np.zeros((m, m), dtype=float)
        price_matrix = np.zeros((m, m), dtype=float)
        delta_matrix = np.zeros((m, m), dtype=float)

        for i in range(K_array.size):
            for j in range(m):
                implied_matrix[i][j] = self.implied_vol_vv(F[j], K_array[i], T[j], K25DeltaPut[j], KATM[j],
                                                        K25DeltaCall[j], sig25DeltaPut[j], sigATM[j], sig25DeltaCall[j], xd[j], xf[j], j)
                price_matrix[i][j] = self.implied_price(F[j], K_array[i], T[j], K25DeltaPut[j], KATM[j],
                                                        K25DeltaCall[j], sig25DeltaPut[j], sigATM[j], sig25DeltaCall[j], xd[j], xf[j], j)
                delta_matrix[i][j] = self.delta(F[j], K_array[i], T[j], xd[j], xf[j], implied_matrix[i][j], j)

        implied_matrix = pd.DataFrame(implied_matrix)
        price_matrix = pd.DataFrame(price_matrix)
        delta_matrix = pd.DataFrame(delta_matrix)

        
        #  Flatten the matrices for use with griddata
        implied_flat = implied_matrix.values.ravel()
        price_flat = price_matrix.values.ravel()
        delta_flat = delta_matrix.values.ravel()


        KK, TT = np.meshgrid(np.array(K_array), np.array(T[:-1]), indexing='ij')

        option_vol = interpolate.griddata((KK.ravel(), TT.ravel()), implied_flat, ([self.t], [self.strike]), method='nearest')
        option_premium = interpolate.griddata((KK.ravel(), TT.ravel()), price_flat, ([self.t], [self.strike]), method='nearest')
        option_delta = interpolate.griddata((KK.ravel(), TT.ravel()), delta_flat, ([self.t], [self.strike]), method='nearest')
   
        #  Implied Volatility vs Strikes at 3M
        fig = plt.figure(figsize = (12, 5))
        X = K_array
        plt.subplot(111)
        plt.xlabel('Strike')
        plt.ylabel('Implied Volatility (%)')
        plt.title('Implied Volatility vs Strikes')
        for i in range(T.size - 1):
            plt.plot(X, 100*implied_matrix[:][i])
        plt.legend(['1D', '1W', '2W', '3W', '1M', '2M', '3M', '6M', '9M', '1Y', '18M', '2Y', '3Y', '7Y', '10Y'], loc = 1)
        plt.show()

        #  Option Premium vs Strikes at 3M
        fig = plt.figure(figsize = (12, 5))
        X = K_array
        plt.subplot(111)
        plt.xlabel('Strike')
        plt.ylabel('C(K) (USD)')
        plt.title('Option Premium vs Strikes')
        for i in range(T.size - 1):
            plt.plot(X, price_matrix[:][i])
        plt.legend(['1D', '1W', '2W', '3W', '1M', '2M', '3M', '6M', '9M', '1Y', '18M', '2Y', '3Y', '7Y', '10Y'])
        plt.show()
        
        return True
    
    def FXOptionPrice(self):

        xlsx_file = pd.ExcelFile(self.importfile)
        vol_data = xlsx_file.parse('Volatility')

        # US Deposit Data
        xd_data = xlsx_file.parse('US_Depo')
        xd = np.array(xd_data['Value'])
        T_xd = pd.DataFrame({'X': xd_data['Term']})

        # EU Deposit Data
        xf_data = xlsx_file.parse('EU_Depo')
        xf = np.array(xf_data['Value'])
        T_xf = pd.DataFrame({'X': xf_data['Term']})

        T = pd.DataFrame({'X': vol_data['Term'].unique()})  # terms

        # dictionary for excel tabulated terms converted to day/360 format
        super_dict = [('M', '*30'), ('D', '*1'), ('W', '*7'), ('Y', '*360'), (' ', '')]
        super_dict = dict(super_dict)

        # Modify the terms to daycount and store the vols in numpy arrays
        T = T.replace(super_dict, regex=True)
        T_xd = T_xd.replace(super_dict, regex=True)
        T_xf = T_xf.replace(super_dict, regex=True)

        T = T.X.apply(lambda x: eval(str(x))) / self.dc
        T_xd = T_xd.X.apply(lambda x: eval(str(x))) / self.dc
        T_xf = T_xf.X.apply(lambda x: eval(str(x))) / self.dc

        sigATM = np.array(vol_data[vol_data['Strike'] == 'ATM']['Value']) / 100  # ATM volatilities
        sig25DeltaRR = np.array(vol_data[vol_data['Strike'] == '25 Delta RR']['Value']) / 100  # ATM volatilities
        sig25DeltaBF = np.array(vol_data[vol_data['Strike'] == '25 Delta BF']['Value']) / 100  # ATM volatilities
        sig25DeltaCall = sig25DeltaBF + sigATM + sig25DeltaRR / 2
        sig25DeltaPut = sig25DeltaBF + sigATM - sig25DeltaRR / 2

        xd, xf = self.rate(xf, xd, vol_data)
        K25DeltaPut, KATM, K25DeltaCall, F = self.delta25_strike(T, sig25DeltaPut, sigATM, sig25DeltaCall, xd, xf)
        option_data = self.create_plots(F, T, K25DeltaPut, KATM, K25DeltaCall, sig25DeltaPut, sigATM, sig25DeltaCall, xd, xf)

        return option_data

def main():

    validus_answer = Vanna_Volga(1.0977, 1.1, 10, 'vanna-volga.xlsx').FXOptionPrice()
    print(validus_answer)
    plt.show()

if __name__ == "__main__":
    main()
```

- **volotility forward skew**: Compared with <em>ITM call</em> and <em>OTM put</em>，the needs for <em>OTM call</em> and <em>ITM put</em> are bigger. So, <em>ITM call</em> and <em>OTM put</em> have smaller vol, while <em>OTM call</em>and <em>ITM put</em>have bigger vol.

- **volotility reverse skew**: <em>ITM call</em> and <em>OTM put</em> are more expensive. So <em>ITM call</em> and <em>OTM put</em> have bigger vol, while <em>OTM call</em> and <em>ITM put</em>have smaller vol.
